# 코딩 전 생각

```python
""""""""""
백트래킹 알고리즘을 활용해서 해결해보자.

def backtrack(now, m):
    if len(now) == m:
        answer.append(now)
        return
        
for i in range(now + 1, n + 1):
    backtrack(now + [i])
    
for i in answer:
    for j in i:
        print(j, end = ' ')
    print()
""""""""""
```

```python
n, m = map(int, input().split())
answer = []
def backtrack(now):
    if len(now) == m:
        answer.append(now)
        return
    
    for i in sorted(list(set(range(1, n+1)) - set(now))):
        backtrack(now + [i])
               
for i in range(1, n+1):
    backtrack([i])
    
for i in answer:
    print(' '.join(map(str, i)))
```
