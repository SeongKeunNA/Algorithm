# 코딩 전 생각

```python
""""""""""
6이3(6666포함) 666 1666 2666 3666 4666 5666 6660 6661 6662 6663 6664 6665 6666 6667 6668 6669 
6이4(6666미포함, 66666포함) 16666 26666 36666 46666 56666 66660 66661 66662 66663 66664 66665 66666 66667 66668 66669
6이6(66666미포함, 666666포함)...

즉, 처음에만 666이고 나머지는 6이 최소 i개 포함되어 있을 때 15개씩이다.

n이 1이면 666이고, 2 <= n < 7이면 str(n - 1) + '666'이고, 7 <= n < 17이면 '666' + str(n-7)이다.


그렇지 않다면(n >= 17), 중간에 들어갈 수는 '6' * (((n - 2) // 15) + 3)이고

만약 (n - 2) % 15(= a)가 5보다 작다면, 앞에 a + 1를 붙인다.

그렇지 않다면, 뒤에 a-5를 붙인다.

""""""""""
```

# 내 답안

```python
n = int(input())
if n == 1: 
    print(666)
else:
    middle ='6' * (((n - 2) // 15) + 3)
    remainder = (n - 2) % 15
    if remainder < 5:
        print(str(remainder+1) + middle)
    else:
        print(middle + str(remainder-5))
```

# 내 답안(2)

```python
n = int(input())
i = '1'
idx = 2
ck = 1
if n == 1:
    print(666)
else:   
    while ck == 1:
        if i[-1] == '6':
            cnt6 = 1
            for j in range(len(i)-2, -1, -1):
                if i[j] == '6':
                    cnt6 +=1
                else:
                    break
            for k in range(0, 10 ** (cnt6)):
                value = i + '6' * (3 - cnt6) + ('0' * (cnt6 - len(str(k)))) + str(k)
                if idx == n:
                    print(value)
                    ck = 0
                    break
                idx += 1

        else:
            value = i + '666'
            if idx == n:
                print(value)
                break
            idx += 1

        i = str(int(i) + 1)
```
