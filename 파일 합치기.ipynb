# 내 답안

```python
import heapq
for t in range(int(input())):
    k = int(input())
    h = list(map(int, input().split()))
    heapq.heapify(h)
    cost = 0
    while True:
        print('h:', h, 'cost:', cost)
        first = heapq.heappop(h)
        second = heapq.heappop(h)
        print('first:', first, 'second:',second)
        heapq.heappush(h, first + second)
        cost += first + second
        print('h:', h, 'cost:',cost)
        print('----------------')
        if len(h) == 1:
            print(cost)
            break 
```

```python
for t in range(int(input())):
    k = int(input())
    files = list(map(int, input().split()))
    dic = dict()
    ls = list(range(k))
    for i in range(0, k):
        dic[str(i)] = files[i]
        
    for i in range(2, k+1): # 1글자, 2글자... k글자 abcd 
        for j in range(k-i+1):
            sub = ls[j: j+i]
            a = dic[' '.join(map(str, sub[1:]))] + files[j] 
            b = dic[' '.join(map(str, sub[:-1]))] + files[j+i-1]
            dic[' '.join(map(str, sub))] = min(a, b)
            
    print(dic[' '.join(map(str, ls))])
```

# 내 답안(6)

```python
from sys import stdin
input = stdin.readline
for t in range(int(input())):
    k = int(input()) #4
    dp = [[0] * (k + 1) for _ in range(k + 1)] # 행까지 글자에서 열까지 글자 가는 것
    files = list(map(int, input().split()))
    sum_value = [0] * (k + 1)
    for i in range(1, k+1):
        sum_value[i] = sum_value[i-1] + files[i-1] 

    for i in range(2, k+1): # 총 i개의 파일을 합침 3
        for j in range(1, k-i+2): # 시작 파일의 번호 1 2
            value = int(1e9)
            for l in range(j, j+i-1): # l번째 파일까지가 앞파일  1
                front = dp[j][l] + sum_value[l] - sum_value[j-1] # 40 30 30 50 / 40 70 100 150 //40
                back = dp[l+1][i+j-1] + sum_value[i+j-1] - sum_value[l] #/ / 60 + 60
                if front+ back < value:
                    value = front+ back
            dp[j][i+j-1] = value
    print(dp[1][k])
```
